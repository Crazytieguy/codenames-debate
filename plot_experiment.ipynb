```python
import pandas as pd

df = pd.read_csv("data/negligent-biased-experiment-2.csv")
df_debate = pd.read_csv("data/negligent-biased-debate-experiment-2.csv")
```

```python
df['mean_exploitation'] = df.mean_expected_score / df.mean_ground_truth_score
df_debate['mean_exploitation'] = df_debate.mean_expected_score / df_debate.mean_ground_truth_score
```

```python
import matplotlib.pyplot as plt


fig, ax = plt.subplots()

df.loc[(df.neglect_words == 1) & (df.bias_neglected_words == 1) & (df.bias_non_neglected_words == 1) & (df.bias_factor == 1.5)].plot(x="adversarial_alpha", y="mean_exploitation", ax = ax)
```
